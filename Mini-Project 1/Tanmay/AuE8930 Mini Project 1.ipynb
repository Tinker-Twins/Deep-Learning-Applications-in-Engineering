### 2.3 Dropout Matrix
Write a function `dropout()` that will randomly set matrix entries to 0 with a given probability. For example, given a 14 by 14 matrix and p=0.2, then 39 entries should be set to 0 at random locations. All other entries will not change. 

**Hint:** $n*m*p$ might not be a integer, you should convert it to an integer using `int()`.

```python
import numpy as np
import random

def dropout(X, p):
    """
        Given an n by m matrix X and a probability p, pick n*m*p entries randomly and set them to 0.
        
        Parameters
        ----------
        X: ndarray
            Input of the given matrix.
        p: float
            Fraction of the matrix entries to drop.

        Return
        ----------
        res: ndarray
            Output of dropout
    """
    # TODO: write your code here
    n = X.shape[0] # Number of rows
    m = X.shape[1] # Number of columns
    flat_array = X.reshape(1, n*m) # Flatten the input matrix (easier for indexing)
    dropout_idx = random.sample(range(len(flat_array[0])), int(n*m*p)) # Generate n*m*p random indices in range of the flattened array size
    for idx in dropout_idx:
        flat_array[0][idx] = 0 # Set values at the random indices to zero
    return flat_array.reshape(n, m) # Return the reshaped n by m matrix

# test code
X = np.random.randint(2, 8, (15, 15))
n = X.shape[0]
m = X.shape[1]
p = 0.15

res = dropout(X, p)
if np.sum(res == 0) == int(n*m*p):
    print("Great")
```

### 2.4 Mini-batch
Creating small batches of training data is a critical step for mini-batch gradient descent. In this exercise, you should implement the function `create_batches(X, y, batch_size)` to generate mini-batches according to a specified batch size. **Note**: the given data should be shuffled in your function before splitting.

For example, the given training data $X$ is a $n\times m$ matrix where $n$ is the number of features and $m$ is the number of samples, and $y$ is a $k\times m$ matrix where $k$ is the dimension of the label and $m$ is the number of samples. The function you created should return two lists to save the splitted $X$ and $y$ respectively. The number of items for each returned list should be the ceiling of $\frac{m}{\text{batch_size}}$.

```python
import numpy as np
from sklearn.utils import shuffle

def create_batches(X, y, batch_size):
    """
        Split the data X and y with given batch size.

        Parameters
        ----------
        X: ndarray
            Features of the training data with shape n by m.
        y: ndarray
            Labels of the training data with shape k by m.
        batch_size: int
            Batch size.

        Return
        ----------
        batches_X: list
            Created small batches of X.
        batches_y: list
            Created small batches of y.
    """
    # TODO: write your code here
    n = X.shape[0] # Number of rows of X
    k = y.shape[0] # Number of rows of y
    m = X.shape[1] # Number of columns of X, y

    X_shuffle, y_shuffle = shuffle(X.T, y.T) # Randomly shuffle data
    data_size = m - m%batch_size # Data size divisible by mini-batches
    batches_X = [] # Initialize X-batches
    batches_y = [] # Initialize y-batches
    for offset in range(0, data_size, batch_size): # Loop through the data with an increment of `batch_size`
        batches_X.append(np.asarray(X_shuffle[offset:offset+batch_size]).T) # X-batches
        batches_y.append(np.asarray(y_shuffle[offset:offset+batch_size]).T) # y-batches
    last_batch_X = np.concatenate((np.asarray(X_shuffle[offset+batch_size:]), np.asarray(X_shuffle[0:batch_size-(len(X_shuffle)-(offset+batch_size))]))) # Concatenate X-batch till end of X_shuffle and remaining X-batch from start of X_shuffle
    last_batch_y = np.concatenate((np.asarray(y_shuffle[offset+batch_size:]), np.asarray(y_shuffle[0:batch_size-(len(y_shuffle)-(offset+batch_size))]))) # Concatenate y-batch till end of X_shuffle and emaining y-batch from start of y_shuffle
    batches_X.append(last_batch_X.T) # X-batches
    batches_y.append(last_batch_y.T) # y-batches
    return batches_X, batches_y

# test code
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Load mnist dataset
(X, y), _ = mnist.load_data()

# Reshape inputs
X = X.reshape(X.shape[0], X.shape[1] * X.shape[2]).T

# Convert labels to one-hot vectors
y = to_categorical(y, 10).T

batch_size = 128
batches_X, batches_y = create_batches(X, y, batch_size)

if isinstance(batches_X, list) and isinstance(batches_y, list):
    if len(batches_X) == 469 and len(batches_y) == 469:
        if batches_X[-1].shape[0] == X.shape[0] and batches_X[0].shape[1] == batch_size:
            print('Great')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# TODO: read dataset
dataset = pd.read_csv('gdrive/My Drive/AuE-8930 Deep Learning/Mini-Project 1/covid_confirmed_SC.csv', usecols=range(1,729)) # Read the CSV file

# TODO: data preprocessing
data_np = dataset.to_numpy() # Convert to numpy array
daily_total = np.sum(data_np, axis=0) # Sum data across all counties
daily_increase = np.diff(daily_total) # Calculate daily increase in cases

# TODO: generate the positive increase number bar plot
x = range(len(daily_increase)) # X-axis data (date integers, 1...729)
plt.figure(figsize=(15, 10)) # Figure size in inches
plt.bar(x, daily_increase, color='red') # Plot daily increase in cases w.r.t. date
plt.title('COVID-19 Analysis') # Plot title
plt.xlabel('Date') # Show label for X-axis
plt.ylabel('Daily Increase in Cases') # Show label for Y-axis
plt.grid(True) # Show grid
plt.show() # Show plot

# TODO: find mean and variance 
print("Mean: {}".format(np.mean(daily_increase))) # Calculate mean of daily increase in cases
print("Variance: {}".format(np.var(daily_increase))) # Calculate variance of daily increase in cases

# TODO: define the function moving_avg(data, n) and find the moving average for 7 days
def moving_avg(data, n):
    """
        Given data and window size, calculate the moving average of data over
        sliding the window.
        
        Parameters
        ----------
        data: ndarray
            Given data

        n: int
            Window size

        Return
        ----------
        moving_avg: list
            Moving averages
    """
    moving_avg = [] # Empty list to store moving averages
    for offset in range(len(data)-n+1): # Loop through the data with an increment of window size
        moving_avg.append(np.mean(data[offset:offset+n])) # Compute moving average per window
    return moving_avg

SMA = moving_avg(daily_increase, 7) # Calculate moving average of daily_increase data over a window of 7 days

# TODO: define the function EWMA(data, alpha) and calculate the exponential weighted moving average when alpha=0.9
def EWMA(data, alpha):
    """
        Given data and smoothening factor, calculate the exponentially weighted
        moving average of the data.
        
        Parameters
        ----------
        data: ndarray
            Given data

        alpha: float
            Smoothening factor between 0 (more smooth) and 1 (less smooth)

        Return
        ----------
        exp_avg: list
            Exponentially weighted moving averages
    """
    exp_avg = [] # Empty list to store exponential moving averages
    exp_avg.append(data[0])
    for i in range(len(data)): # Loop through the data
        exp_avg.append((alpha*data[i]) + (1-alpha)*exp_avg[-1]) # Compute exponential moving average per data point
    return exp_avg[1:]

EMA = EWMA(daily_increase, 0.9) # Calculate exponential moving average of daily_increase data with alpha = 0.9

# TODO: plot the daily increase cases curve for moving average results and exponential weighted moving average results
plt.figure(figsize=(15, 10)) # Figure size in inches
plt.bar(np.linspace(1, len(daily_increase), len(daily_increase)), daily_increase, label='Raw Data', color='red') # Plot raw data (daily increase in cases)
plt.plot(np.linspace(1, len(SMA), len(SMA)), SMA, label='Simple Moving Average (SMA)', color='blue') # Plot simple moving average of daily increase in cases
plt.plot(np.linspace(1, len(EMA), len(EMA)), EMA, label='Exponentially Weighted Moving Average (EWMA)', color='green') # Plot exponential moving average of daily increase in cases
plt.title('COVID-19 Analysis')
plt.legend(loc='upper left') # Show legend
plt.xlabel('Date') # Show label for X-axis
plt.ylabel('Daily Increase in Cases') # Show label for Y-axis
plt.grid(True) # Show grid
plt.show() # Show plot
```
