### 2.1 Matrix basics
Given: 
$$ A = 
 \left[
 \begin{matrix}
   1 & 2 & 3 \\
   4 & 5 & 6 \\
   7 & 8 & 9
  \end{matrix}
  \right]
$$

$$
B = 
 \left[
 \begin{matrix}
   1 & 2 & 3 \\
   4 & 5 & 6 \\
   7 & 8 & 9
  \end{matrix}
  \right]
$$

$$ C_1 = [1, 2, 3], C_2 = [4, 5, 6], C_3 = [7, 8, 9]$$

1. Compute Hadamard product (element-wise product) of $A$ and $B$

2. Compute dot product of $A$ and $B$

3. Compute Frobenius Norm of $A$ (use loop)

4. Compute the sum of all the elements of $A$ by using np.sum()

5. Concatenate $C_1, C_2, C_3$ as a new matrix $C$ with shape of 3 by 3 by using np.concatenate() and np.vstack() 

6. Reshape $C$ into a 1 by 9 vector by using np.reshape()



```python
A = np.arange(1, 10, 1).reshape(3, 3)
B = np.arange(1, 10, 1).reshape(3, 3)
C1 = np.copy(A[0])
C2 = np.copy(A[1])
C3 = np.copy(A[2])

# TODO: element-wise product of A and B
hadamard_product = np.multiply(A,B) # compute element-wise product of A and B
print("Hadamard product (element-wise product) of A and B matrices:\n",hadamard_product,"\n")

# TODO: dot product of A and B
dot_product = np.dot(A,B) # compute dot product of A and B
print("Dot product of A and B matrices:\n",dot_product,"\n")

# TODO: Frobenius norm of A (use loop)
# compute Frobenius norm of A (using numpy)
# frobenius_norm = np.linalg.norm(A)
# compute Frobenius norm of A (using loop)
sum_of_squares = 0
for rows in range(A.shape[0]):
    for columns in range(A.shape[1]):
        sum_of_squares += A[rows][columns]**2
frobenius_norm = np.sqrt(sum_of_squares)
print("Frobenius norm of A matrix:\n",round(frobenius_norm,4),"\n")

# TODO: compute the sum of all entries of A
elements_sum = np.sum(A) # compute sum of all elements of A
print("Sum of all elements of A matrix:\n",elements_sum,"\n")

# TODO: concatenate
C_concatenate = np.concatenate((C1, C2, C3)) # concatenate row vectors into a single row vector
C_concatenate = C_concatenate.reshape(3, 3) # reshape row vector into a matrix of specified dimensions
print("C matrix from C1, C2, C3 vectors using np.concatenate():\n",C_concatenate,"\n")
C_vstack = np.vstack((C1, C2, C3)) # stack row vectors vertically to form a matrix
print("C matrix from C1, C2, C3 vectors using np.vstack():\n",C_vstack,"\n")

# TODO: reshape
C_reshaped = np.reshape(C_vstack, (1, 9))
print("C matrix reshaped into 1 by 9 vector:\n",C_reshaped,"\n")
```

### 2.3 Dropout Matrix
Write a function `dropout()` that will randomly set matrix entries to 0 with a given probability. For example, given a 14 by 14 matrix and p=0.2, then 39 entries should be set to 0 at random locations. All other entries will not change. 

**Hint:** $n*m*p$ might not be a integer, you should convert it to an integer using `int()`.

```python
def dropout(X, p):
    """
        Given an n by m matrix X and a probability p, pick n*m*p entries randomly and set them to 0.
        
        Parameters
        ----------
        X: ndarray
            Input of the given matrix.
        p: float
            Fraction of the matrix entries to drop.

        Return
        ----------
        res: ndarray
             Output of dropout
    """
    # TODO: write your code here
    n = X.shape[0] # number of rows
    m = X.shape[1] # number of columns
    number_entries = int(n*m*p) # number of elements to replace based on input matrix size and dropout probability
    X = X.flatten() # flatten input matrix into a row vector
    replacement_indices = np.random.choice(range(X.shape[0]), number_entries, replace=False) # select random entries
    X[replacement_indices] = 0 # set random entries to 0
    X = X.reshape(n,m) # reshape (flattened) row vector to matrix of original dimensions
    return X

# test code
X = np.random.randint(2, 8, (15, 15))
n = X.shape[0] # number of rows
m = X.shape[1] # number of columns
p = 0.15
res = dropout(X, p)
if np.sum(res == 0) == int(n*m*p):
    print("Great")
```

### 2.4 Mini-batch
Creating small batches of training data is a critical step for mini-batch gradient descent. In this exercise, you should implement the function `create_batches(X, y, batch_size)` to generate mini-batches according to a specified batch size. **Note**: the given data should be shuffled in your function before splitting.

For example, the given training data $X$ is a $n\times m$ matrix where $n$ is the number of features and $m$ is the number of samples, and $y$ is a $k\times m$ matrix where $k$ is the dimension of the label and $m$ is the number of samples. The function you created should return two lists to save the splitted $X$ and $y$ respectively. The number of items for each returned list should be the ceiling of $\frac{m}{\text{batch_size}}$.

```python
def create_batches(X, y, batch_size):
    """
        Split the data X and y with given batch size.

        Parameters
        ----------
        X: ndarray
            Features of the training data with shape n by m.
        y: ndarray
            Labels of the training data with shape k by m.
        batch_size: int
            Batch size.

        Return
        ----------
        batches_X: list
            Created small batches of X.
        batches_y: list
            Created small batches of y.
    """
    # TODO: write your code here
    X_shuffled = X.T # transpose array of features
    y_shuffled = y.T # transpose array of labels
    np.random.seed(22) # set a constant seed for randomization (for generating pseudo-random numbers)
    np.random.shuffle(X_shuffled) # shuffle features (with constant seed)
    np.random.shuffle(y_shuffled) # shuffle labels (with constant seed)
    batches_X = [] # initialize list for batches of features
    batches_y = [] # initialize list for batches of labels
    no_of_batches = int(np.ceil(X.shape[1]/batch_size)) # number of batches based on number of samples and batch size
    for i in range(no_of_batches): # iterate over all mini batches
        batch_X = X_shuffled[i*batch_size:(i+1)*batch_size] # create mini batch for features
        batch_y = y_shuffled[i*batch_size:(i+1)*batch_size] # create mini batch for labels
        batches_X.append(np.asarray(batch_X).T) # list of numpy arrays containing all mini batch for features
        batches_y.append(np.asarray(batch_y).T) # list of numpy arrays containing all mini batch for labels
    return batches_X, batches_y

# test code
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Load mnist dataset
(X, y), _ = mnist.load_data()

# Reshape inputs
X = X.reshape(X.shape[0], X.shape[1] * X.shape[2]).T

# Convert labels to one-hot vectors
y = to_categorical(y, 10).T

batch_size = 128
batches_X, batches_y = create_batches(X, y, batch_size)

if isinstance(batches_X, list) and isinstance(batches_y, list):
    if len(batches_X) == 469 and len(batches_y) == 469:
        if batches_X[-1].shape[0] == X.shape[0] and batches_X[0].shape[1] == batch_size:
            print('Great')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
import pandas as pd
%matplotlib inline

# TODO: read dataset
drive.mount("/content/drive") # mount the drive
df = pd.read_csv("/content/drive/MyDrive/Coursework/Deep Learning/Mini Project 01/COVID_Confirmed_SC.csv") # read the csv file
#print(df.head()) # print out first 5 samples from the dataframe

# TODO: data preprocessing
df.drop(columns=df.columns[0], axis=1, inplace=True) # remove first column (county names) of dataframe
daily_cumulative = df.sum() # daily cumulative increase of positive patients across all counties
total_daily_increase = daily_cumulative.diff() # total daily increase in positive patients across all counties
print("\nTotal daily increase data:\n",total_daily_increase)

# TODO: generate the positive increase number bar plot
# define figure size
plt.figure(figsize=(25, 10))
# plot positive increase number
x = np.linspace(1, total_daily_increase.size, total_daily_increase.size)
y = total_daily_increase.to_numpy()
plt.bar(x[1:], y[1:], color="blue", linewidth=2, linestyle='-')
# add title
plt.title("SC COVID-19 Confirmed")
# add axes labels
plt.ylabel('Daily Increase in Cases')
plt.xlabel('Day Index')
# add grid
plt.grid()

# TODO: find mean and variance 
mean = np.mean(total_daily_increase.to_numpy()[1:]) # compute mean
print("\nMean of data:",mean,"\n")
variance = np.var(total_daily_increase.to_numpy()[1:]) # compute variance
print("Variance of data:",variance,"\n")

# TODO: define the function moving_avg(data, n) and find the moving average for 7 days
def moving_avg(data, n):
    """
        Find the moving average of data over n samples.

        Parameters
        ----------
        data: ndarray
            Input data.
        n: int
            Number of samples to use for finding moving average.

        Return
        ----------
        moving_avg: ndarray
            Moving average of data over n samples.
    """
    moving_avg = [] # Initialize list to store moving averages of data
    i = 0 # iteration index
    while i < len(data) - n+1: # loop through data
        window = data[i : i+n] # current window
        window_avg = sum(window)/n # current window average
        moving_avg.append(window_avg) # add window average to moving averages list
        i += 1 # Shift window by one sample
    return moving_avg

moving_avg = moving_avg(total_daily_increase.to_numpy()[1:],7)
print("Moving average of data:",moving_avg,"\n")

# TODO: define the function EWMA(data, alpha) and calculate the exponential weighted moving average when alpha=0.9
def EWMA(data, alpha):
    """
        Find the exponential weighted moving average of data.

        Parameters
        ----------
        data: ndarray
            Input data.
        alpha: float
            Weight or smoothening factor.

        Return
        ----------
        ewma: ndarray
            Exponential weighted moving average of data.
    """
    
    ewma = [] # Initialize list to store exponential weighted moving averages of data
    ewma.append(data[0]) # first ewma for recursion
    i = 1 # iteration index
    while i < len(data): # loop through data
        window_avg = (alpha*data[i])+(1-alpha)*ewma[-1] # emwa recursive function (current window)
        ewma.append(window_avg) # add window average to moving averages list
        i += 1 # Shift window by one sample
    return ewma

ewma = EWMA(total_daily_increase.to_numpy()[1:],0.9)
print("Exponential weighted moving average of data:",moving_avg,"\n")

# TODO: plot the daily increase cases curve for moving average results and exponential weighted moving average results
# define figure size
plt.figure(figsize=(25, 10))
# plot positive increase number
x = np.linspace(1, total_daily_increase.size, total_daily_increase.size)
y = total_daily_increase.to_numpy()
plt.bar(x[1:], y[1:], color="blue", label="Raw", linewidth=2, linestyle='-') # plot raw data (bar plot)
plt.plot(x[1:len(moving_avg)+1], moving_avg, color="red", label="MA", linewidth=2, linestyle='--') # plot moving average of data
plt.plot(x[1:len(ewma)+1], ewma, color="gray", label="EWMA", linewidth=2, linestyle='-.') # plot exponential weighted moving average of data
# add title
plt.title("SC COVID-19 Confirmed")
# add legend
plt.legend()
# add axes labels
plt.ylabel('Daily Increase in Cases')
plt.xlabel('Day Index')
# add grid
plt.grid()
```
