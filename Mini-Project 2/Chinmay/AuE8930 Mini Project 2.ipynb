$$X_{\text{scaled}}=X_{\text{std}}\cdot(b-a) + a$$

### Task 2.1: Initialize model parameters
Since the model is a two-layer (one hidden layer and an output layer) neural network, model parameters will be $W_1$, $b_1$, $W_2$, and $b_2$. In this task, you should implement a function of `init_params` to initialize parameters based on [He method](https://towardsdatascience.com/weight-initialization-techniques-in-neural-networks-26c649eb3b78).

**Hints:**
- $W_1$ should be an $l_1\times l_0$ matrix where $l_1$ is the hidden layer size and $l_0$ is the feaure dimension of input samples.
- $b_1$ should be an $l_1\times1$ matrix.
- $W_2$ and $b_2$ should have the silimar dimensionality.
- Use `np.random.randn()` to generate random values.

```python
def init_params(input_layer_size, hidden_layer_size, output_layer_size):
    """
    Initialize model parameters based on He initialization method.
        
    Parameters
    ----------
    input_layer_size: int
        Number of features in the input sample.
    hidden_layer_size: int
        Number of neurons in the hidden layer.
    output_layer_size: int
        Number of neurons in the output layer.

    Return
    ----------
    W1: ndarray
        Output of initialized W1.
    b1: ndarray
        Output of initialized b1.
    W2: ndarray
        Output of initialized W2.
    b2: ndarray
        Output of initialized b2.
    """
    # TODO: write your code here
    W1 = np.random.randn(hidden_layer_size,input_layer_size)*np.sqrt(2/(input_layer_size)) # weights for first layer
    b1 = np.zeros((hidden_layer_size,1)) # biases for first layer
    W2 = np.random.randn(output_layer_size,hidden_layer_size)*np.sqrt(2/(hidden_layer_size)) # weights for second layer
    b2 = np.zeros((output_layer_size,1)) # biases for second layer
    return W1, b1, W2, b2
```

### Task 2.2: Implement ReLU activation function
ReLU is the most common activation function used in neural networks. Implement the ReLU function and use it in your model.
$$a = \begin{cases}
   z &\text{if } z>0 \\
   0 &\text{if } z\le0
\end{cases}$$

$$\frac{da}{dz} = \begin{cases}
   1 &\text{if } z>0 \\
   0 &\text{if } z\le0
\end{cases}$$

```python
def relu(Z, derivative=False):
    """
    If derivative=False, return the result transformed by the ReLU function.
    If derivative=True, return the result transformed by the derivative of the ReLU function.

    Parameters
    ----------
    Z: ndarray
        Input of the given data.
    derivative: bool.
        Whether or not to take the derivative of the ReLU function.

    Return
    ----------
    A: ndarray
        Output of the transformed matrix.
    """
    # TODO: write your code here
    if derivative: # if derivative flag is set
        if Z.all() > 0:
            A = np.ones(Z.shape) # compute result transformed by the derivative of the ReLU function
        else:
            A = np.zeros(Z.shape) # compute result transformed by the derivative of the ReLU function
    else:
        if Z.all() > 0:
            A = Z # compute result transformed by the ReLU function
        else:
            A = np.zeros(Z.shape) # compute result transformed by the ReLU function
    return A
```

### Task 2.3: Implement forward propagation function
The forward function is used to define how your model is going to be run, from input to output.

**Hints:**
- Computations in the forward propagation:
$$Z^{[1]}=W^{[1]}X+b^{[1]}$$
$$A^{[1]}=g(Z^{[1]})$$
$$Z^{[2]}=W^{[2]}A^{[1]}+b^{[2]}$$
$$\hat{y}=g(Z^{[2]})$$
where $g()$ is the ReLU activation function.

```python
def forward(X, W1, b1, W2, b2):
    """
    Generates output predictions for the input samples with given params.
        
    Parameters
    ----------
    X: ndarray
        Input samples with shape of n x m where n is the feature dimension 
        and m is the number of samples.
    W1: ndarray
        Weights of the hidden layer.
    b1: ndarray
        Intercepts of the hidden layer.
    W2: ndarray
        Weights of the output layer.
    b2: ndarray
        Intercepts of the output layer.

    Return
    ----------
    Z1: ndarray
        Output of the hidden layer before applying activation function.
    A1: ndarray
        Output of the hidden layer
    Z2: ndarray
        Output of the output layer before applying activation function.
    A2: ndarray
        Output of the output layer.
    """
    # TODO: write your code here
    Z1 = np.matmul(W1, X) + b1 # compute output of first layer before applying activation function
    A1 = relu(Z1, derivative=False) # compute output of first layer after applying activation function
    Z2 = np.matmul(W2, A1) + b2 # compute output of second layer before applying activation function
    A2 = relu(Z2, derivative=False) # compute output of second layer after applying activation function
    return Z1, A1, Z2, A2
```

### Task 2.4: Implement loss function
For this regression problem, we will use mean squared error (MSE) as the loss function.
$$J=\frac{1}{2m}\sum_{i=1}^m\left(\hat{y}^{(i)} - y^{(i)}\right)^2$$

```python
def loss(y_truth, y_pred):
    """
    Return the mean squared error between the ground truth and predictions.
        
    Parameters
    ----------
    y_truth: ndarray
        An 1 x m matrix of ground truth.
    y_pred: ndarray
        An 1 x m matrix of predictions.

    Return
    ----------
    loss: float
        Output of the mean squared error.
    """
    # TODO: write your code here
    loss = 0.5 * np.square(np.subtract(y_pred, y_truth)).mean() # MSE loss function
    return loss
```

### Task 2.5: Implement backpropagation function
The backward calculates the gradient of the loss function with respect to the neural network's weights and intercepts.


```python
def backward(X, y, Z1, A1, Z2, A2, W1, b1, W2, b2):
    """
    Return the gradient of the loss function with respect to the model's weights.
        
    Parameters
    ----------
    X: ndarray
        Input samples with shape of n x m where n is the feature dimension 
        and m is the number of samples.
    y: ndarray
        An 1 x m matrix of ground truth.
    Z1: ndarray
        Output of the hidden layer before applying activation function.
    A1: ndarray
        Output of the hidden layer.
    Z2: ndarray
        Output of the output layer before applying activation function.
    A2: ndarry
        Output of the output layer.
    W1: ndarray
        Weights of the hidden layer.
    b1: ndarray
        Intercepts of the hidden layer.
    W2: ndarray
        Weights of the output layer.
    b2: ndarray
        Intercepts of the output layer.

    Return
    ----------
    dW1: ndarray
        Gradients of weiths at the hidden layer.
    db1: ndarray
        Gradients of intercepts at the hidden layer.
    dW2: ndarray
        Gradients of weiths at the output layer.
    db2: ndarray
        Gradients of intercepts at the output layer.
    """
    # TODO: write your code here
    m = X.shape[1] # number of samples
    dZ2 = A2 - y # dZ computation for second layer OURTODO: check this
    dW2 = 1/m * np.matmul(dZ2, A1.T) # dW computation for second layer using dZ computation
    db2 = 1/m * np.sum(dZ2, axis=1, keepdims=True) # db computation for second layer using dZ computation
    dZ1 = np.multiply(np.matmul(W2.T, dZ2), relu(Z1, derivative=True)) # dZ computation for first layer
    dW1 = 1/m * np.matmul(dZ1, X.T) # dW computation for first layer using dZ computation
    db1 = 1/m * np.sum(dZ1, axis=1, keepdims=True) # db computation for first layer using dZ computation
    return dW1, db1, dW2, db2
```

### Task 2.6: Implement update function
In this part, we will update model parameters according to the generated gradients.
$$W:=W-\alpha\cdot dW$$
$$b:=b-\alpha\cdot db$$

```python
def update(lr, W1, b1, W2, b2, dW1, db1, dW2, db2):
    """
    Return updated parameters based on the gradient descent optimization. 
        
    Parameters
    ----------
    lr: float
        Learning rate.
    W1: ndarray
        Weights of the hidden layer.
    b1: ndarray
        Intercepts of the hidden layer.
    W2: ndarray
        Weights of the output layer.
    b2: ndarray
        Intercepts of the output layer.
    dW1: ndarray
        Gradients of weiths at the hidden layer.
    db1: ndarray
        Gradients of intercepts at the hidden layer.
    dW2: ndarray
        Gradients of weiths at the output layer.
    db2: ndarray
        Gradients of intercepts at the output layer.

    Return
    ----------
    W1_: ndarray
        Updated weights of the hidden layer.
    b1_: ndarray
        Updated intercepts of the hidden layer.
    W2_: ndarray
        Updated weights of the output layer.
    b2_: ndarray
        Updated intercepts of the output layer.
    """
    # TODO: write your code here
    W1_ = W1-lr*dW1 # update weights for first layer
    b1_ = b1-lr*db1 # update biases for first layer
    W2_ = W2-lr*dW2 # update weights for second layer
    b2_ = b2-lr*db2 # update biases for second layer
    return W1_, b1_, W2_, b2_
```

### Task 2.7: Implement train function
So far, you have implemented functions that are required to train the neural network model. It's time to organize them and complete the training step. Let's keep moving on!

**Note:** Since you have learned how to create mini-batches in Mini-Project 1, just directly use the given `create_batches()`.

```python
def create_batches(X, y, batch_size):
    """
    Split the data X and y with given batch size.

    Parameters
    ----------
    X: ndarray
        Features with shape n by m.
    y: ndarray
        Labels with shape k by m.
    batch_size: int
        Batch size.

    Return
    ----------
    batches_X: list
        Created small batches of X.
    batches_y: list
        Created small batches of y.
    """
    # Get the number of examples
    m = X.shape[1]
    # Get label dimension
    k = y.shape[0]

    # Init variables to save batches
    batches_X = []
    batches_y = []
    
    # Combine X and y
    data = np.vstack((X, y))
    np.random.shuffle(data.T)
        
    # Calcuate the number of mini batches
    n_minibatches = m // batch_size 

    for i in range(n_minibatches): 
        mini_batch = data[:, i * batch_size:(i + 1) * batch_size]        
        batches_X.append(mini_batch[:-k, :])
        batches_y.append(mini_batch[-k:, :])
        
    # Handle the rest data
    if m % batch_size != 0: 
        mini_batch = data[:, n_minibatches * batch_size:]        
        batches_X.append(mini_batch[:-k, :])
        batches_y.append(mini_batch[-k:, :])
        
    return batches_X, batches_y
```

```python
def train(X_train, y_train, X_test, y_test, W1, b1, W2, b2, lr, batch_size, epochs):
    """
    Train the model based on the mini-batch graient descent.

    Parameters
    ----------
    X_train: ndarray
        Features of the training data with shape n by m1.
    y_train: ndarray
        Labels of the training data with shape 1 by m1.
    X_test: ndarray
        Features of the training data with shape n by m2.
    y_test: ndarray
        Labels of the training data with shape 1 by m2.
    W1: ndarray
        Initial weights of the hidden layer.
    b1: ndarray
        Initial intercepts of the hidden layer.
    W2: ndarray
        Initial weights of the output layer.
    b2: ndarray
        Initial intercepts of the output layer.
    lr: float
        Learning rate.
    batch_size: int
        Batch size.
    epochs: int
        Number of epochs to train the model. An epoch is an iteration over 
        the entire x and y data provided.

    Return
    ----------
    training_loss_history: list
        Each entry stands for the average training loss of all iterations in a certain epoch.
    test_loss_history: list
        A record of test losses.
    """
    # TODO: write your code here
    training_loss_history = np.zeros((epochs, 1)) # initialize training loss history list
    test_loss_history = np.zeros((epochs, 1)) # initialize test loss history list
    for i in range(epochs):
        # create mini-batches
        batches_X, batches_y = create_batches(X_train, y_train, batch_size)
        # training portion
        training_loss = np.zeros((len(batches_X), 1))
        for t in range(len(batches_X)):
            Z1, A1, Z2, A2 = forward(batches_X[t], W1, b1, W2, b2) # forward propagation
            training_loss[t] = loss(batches_y[t], A2) # training loss computation
            dW1, db1, dW2, db2 = backward(batches_X[t], batches_y[t], Z1, A1, Z2, A2, W1, b1, W2, b2) # backward propagation
            W1, b1, W2, b2 = update(lr, W1, b1, W2, b2, dW1, db1, dW2, db2) # parameter updation
        training_loss_history[i] = np.mean(training_loss) # store training loss
        # test portion
        Z1, A1, Z2, A2 = forward(X_test, W1, b1, W2, b2) # forward propagation
        test_loss_history[i] = loss(y_test, A2) # test loss computation and storage
    return training_loss_history, test_loss_history
```

```python
class DNN:
    def __init__(self, input_size, layer_sizes, epochs=100, lr=0.001, batch_size=64):
        """
        Parameters
        ----------
        input_size: int
            Number of features in the inputs.
        layer_sizes: list
            Number of units in each hidden layer and the output. 
        epochs: int
            Number of epochs to train the model
        lr: float
            Learning rate.
        batch_size: int
            Batch size.
        """
        self.input_size = input_size
        self.layer_sizes = layer_sizes
        self.batch_size = batch_size
        self.epochs = epochs
        self.lr = lr

        # Init model parameters
        self.params = self._init_params()


    def _init_params(self):
        """ 
        Initialize model parameters based on He method.

        Return
        ----------
        params: dict
            Initialized model parameters represented by key-value pairs.
        """
        # TODO: write your code here
        params = {} # initialize parameters dictonary

        params['W1'] = np.random.randn(self.layer_sizes[0],self.input_size)*np.sqrt(2/(self.input_size)) # weights for first layer
        params['b1'] = np.zeros((self.layer_sizes[0],1)) # biases for first layer

        for l in range(len(self.layer_sizes)-1):
            params['W'+str(l+2)] = np.random.randn(self.layer_sizes[l+1],self.layer_sizes[l])*np.sqrt(2/(self.layer_sizes[l])) # weights for subsequent layer
            params['b'+str(l+2)] = np.zeros((self.layer_sizes[l+1],1)) # biases for subsequent layer

        return params


    def _relu(self, Z, derivative=False):
        """
        If derivative=False, return the result transformed by the ReLU function.
        If derivative=True, return the result transformed by the derivative of the ReLU function.

        Parameters
        ----------
        Z: ndarray
            Input of the given data.
        derivative: bool.
            Whether or not to take the derivative of the ReLU function.

        Return
        ----------
        A: ndarray
            Output of the transformed matrix.
        """
        # TODO: write your code here
        if derivative: # if derivative flag is set
            if Z.all() > 0:
                A = np.ones(Z.shape) # compute result transformed by the derivative of the ReLU function
            else:
                A = np.zeros(Z.shape) # compute result transformed by the derivative of the ReLU function
        else:
            if Z.all() > 0:
                A = Z # compute result transformed by the ReLU function
            else:
                A = np.zeros(Z.shape) # compute result transformed by the ReLU function

        return A


    def _forward(self, X):
        """
        Generates output predictions for the input samples with given params.
            
        Parameters
        ----------
        X: ndarray
            Input samples with shape of n x m where n is the feature dimension 
            and m is the number of samples.

        Return
        ----------
        output_layers: dict
            Outputs of each layer represented by key-value pairs.
        """
        # TODO: write your code here
        output_layers = {} # initialize output predictions dictonary

        output_layers['Z1'] = np.matmul(self.params.get('W1'), X) + self.params.get('b1') # output of first layer before applying activation function
        output_layers['A1'] = self._relu(output_layers.get("Z1"), derivative=False) # output of first layer after applying activation function

        for l in range(len(self.layer_sizes)-1):
            output_layers['Z'+str(l+2)] = np.matmul(self.params.get("W"+str(l+2)), output_layers.get("A"+str(l+1))) + self.params.get("b"+str(l+2)) # output of l-th layer before applying activation function
            output_layers['A'+str(l+2)] = self._relu(output_layers.get("Z"+str(l+2)), derivative=False) # output of l-th layer after applying activation function

        return output_layers


    @staticmethod
    def mse(y_truth, y_pred):
        """
        Return the mean squared error between the ground truth and predictions.
            
        Parameters
        ----------
        y_truth: ndarray
            An 1 x m matrix of ground truth.
        y_pred: ndarray
            An 1 x m matrix of predictions.

        Return
        ----------
        loss: float
            Output of the mean squared error.
        """
        # TODO: write your code here
        loss = 0.5 * np.square(np.subtract(y_pred, y_truth)).mean() # MSE loss function
        return loss


    def _backward(self, X, y, output_layers):
        """
        Return the gradient of the loss function with respect to the model's weights.
            
        Parameters
        ----------
        X: ndarray
            Input samples with shape of n x m where n is the feature dimension 
            and m is the number of samples.
        y: ndarray
            An 1 x m matrix of ground truth.
        output_layers: dict
            Outputs of each layer represented by key-value pairs.

        Return
        ----------
        gradients: dict
            Gradients of weights and intercepts at each layers.
        """
        # TODO: write your code here
        gradients = {} # initialize gradients dictonary
        m = X.shape[1] # number of samples

        dZL = output_layers.get('A'+str(len(self.layer_sizes))) - y # dZ computation for last layer
        dWL = 1/m * np.matmul(dZL, output_layers.get('A'+str(len(self.layer_sizes)-1)).T) # dW computation for last layer using dZ computation
        dbL = 1/m * np.sum(dZL, axis=1, keepdims=True) # db computation for last layer using dZ computation
        gradients['dZ'+str(len(self.layer_sizes))] = dZL # gradient dZ of last layer
        gradients['dW'+str(len(self.layer_sizes))] = dWL # gradient dW of last layer
        gradients['db'+str(len(self.layer_sizes))] = dbL # gradient dB of last layer

        for l in range(len(self.layer_sizes)-2):
            dZl = np.multiply(np.matmul(self.params.get('W'+str(len(self.layer_sizes)-l)).T, gradients.get('dZ'+str(len(self.layer_sizes)-l))), relu(output_layers.get('Z'+str(len(self.layer_sizes)-(l+1))), derivative=True)) # dZ computation for l-th layer
            dWl = 1/m * np.matmul(dZl, output_layers.get('A'+str(len(self.layer_sizes)-(l+2))).T) # dW computation for l-th layer using dZ computation
            dbl = 1/m * np.sum(dZl, axis=1, keepdims=True) # db computation for l-th layer using dZ computation
            gradients['dZ'+str(len(self.layer_sizes)-(l+1))] = dZl # gradient dZ of l-th layer
            gradients['dW'+str(len(self.layer_sizes)-(l+1))] = dWl # gradient dW of l-th layer
            gradients['db'+str(len(self.layer_sizes)-(l+1))] = dbl # gradient dB of l-th layer

        dZ1 = np.multiply(np.matmul(self.params.get('W'+str(2)).T, gradients.get('dZ'+str(2))), relu(output_layers.get('Z'+str(1)), derivative=True)) # dZ computation for first layer
        dW1 = 1/m * np.matmul(dZ1, X.T) # dW computation for first layer using dZ computation
        db1 = 1/m * np.sum(dZ1, axis=1, keepdims=True) # db computation for first layer using dZ computation
        gradients['dZ'+str(1)] = dZ1 # gradient dZ of first layer
        gradients['dW'+str(1)] = dW1 # gradient dW of first layer
        gradients['db'+str(1)] = db1 # gradient dB of first layer

        return gradients


    def _update(self, gradients):
        """
        Update network parameters according to the update rule of gradient descent. 
            
        Parameters
        ----------
        gradients: dict
            Parameter gradients.
        """
        # TODO: write your code here
        for l in range(len(self.layer_sizes)-1):
            self.params.update({'W'+str(l+1): self.params['W'+str(l+1)] - self.lr * gradients.get('dW'+str(l+1))}) # update weights for l-th layer
            self.params.update({'b'+str(l+1): self.params['b'+str(l+1)] - self.lr * gradients.get('db'+str(l+1))}) # update biases for l-th layer


    def train(self, X_train, y_train, X_test, y_test):
        """
        Train the model based on mini-batch graient descent.

        Parameters
        ----------
        X_train: ndarray
            Features of the training data with shape n by m1.
        y_train: ndarray
            Labels of the training data with shape 1 by m1.
        X_test: ndarray
            Features of the training data with shape n by m2.
        y_test: ndarray
            Labels of the training data with shape 1 by m2.

        Return
        ----------
        training_loss_history: list
            Entries are average training losses of all iterations in one epoch.
        test_loss_history: list
            A record of test loss values.
        """
        # TODO: write your code here
        training_loss_history = np.zeros((self.epochs, 1)) # initialize training loss history list
        test_loss_history = np.zeros((self.epochs, 1)) # initialize test loss history list
        for i in range(self.epochs):
            # create mini-batches
            batches_X, batches_y = create_batches(X_train, y_train, batch_size)
            # training portion
            training_loss = np.zeros((len(batches_X), 1))
            for t in range(len(batches_X)):
                output_layers = self._forward(batches_X[t]) # forward propagation
                training_loss[t] = self.mse(batches_y[t], output_layers.get('A'+str(len(self.layer_sizes)))) # training loss computation
                gradients = self._backward(batches_X[t], batches_y[t], output_layers) # backward propagation
                self._update(gradients) # parameter updation
            training_loss_history[i] = np.mean(training_loss) # store training loss
            # test portion
            output_layers = self._forward(X_test) # forward propagation
            test_loss_history[i] = self.mse(y_test, output_layers.get('A'+str(len(self.layer_sizes)))) # test loss computation and storage
        return training_loss_history, test_loss_history
        
    
    def _create_batches(self, X, y):
        """
            Split X and y into small batches with given batch size.

            Parameters
            ----------
            X: ndarray
                Features with shape n by m.
            y: ndarray
                Labels with shape k by m.

            Return
            ----------
            batches_X: list
                Created small batches of X.
            batches_y: list
                Created small batches of y.
        """
        # Get the number of examples
        m = X.shape[1]
        # Get label dimension
        k = y.shape[0]

        # Init variables to save batches
        batches_X = []
        batches_y = []
        
        # Combine X and y
        data = np.vstack((X, y))
        np.random.shuffle(data.T)
            
        # Calcuate the number of mini batches
        n_minibatches = m // self.batch_size

        for i in range(n_minibatches): 
            mini_batch = data[:, i * self.batch_size:(i + 1) * self.batch_size]        
            batches_X.append(mini_batch[:-k, :])
            batches_y.append(mini_batch[-k:, :])
            
        # Handle the rest data
        if m % self.batch_size != 0: 
            mini_batch = data[:, n_minibatches * self.batch_size:]        
            batches_X.append(mini_batch[:-k, :])
            batches_y.append(mini_batch[-k:, :])
            
        return batches_X, batches_y
```
